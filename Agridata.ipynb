```python
# ------5 ----------------------------------------------------------------------------------------------------
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)
#	select year, SUM(sugarcane_production) AS total_production 
#	FROM AgriInfo GROUP BY year ORDER BY year DESC LIMIT 50 ;
Curryear = df["year"].max()
# Filter data for the last 50 years
df_sugarcane = df[df["year"] >= (Curryear - 50)]
sugarcane_production=df_sugarcane.groupby("year")["sugarcane_production"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x="year", y="sugarcane_production", data=sugarcane_production, marker="o", color="green")
plt.xlabel("Year")
plt.ylabel("Sugarcane Production")
plt.title("India's Sugarcane Production Over Last 50 Years")
plt.show()
```

```python
#Rice Production Vs Wheat Production (Last 50y)R

	#select year, SUM(rice_production) rice_production, SUM(wheat_production) wheat_production
	#FROM AgriInfo GROUP BY year ORDER BY year DESC LIMIT 50;
Curryear = df["year"].max()
df_Rice_wheat = df[df["year"] >= (Curryear - 50)]
Rice_wheat_production = df_Rice_wheat.groupby("year")[["rice_production", "wheat_production"]].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x="year", y="rice_production", data=Rice_wheat_production, marker="o", color="blue", linewidth=2, label="Rice Production")
sns.lineplot(x="year", y="wheat_production", data=Rice_wheat_production, marker="o", color="red", linewidth=2, label="Wheat Production")

# Customize the plot
plt.xlabel("Year")
plt.ylabel("Production (in tons)")
plt.title("Rice vs. Wheat Production Over the Years")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

```

```python
# 6--------------------------------
#Rice Production By West Bengal Districts

	#select district_name, SUM(rice_production) AS total_production 
	#FROM AgriInfo WHERE statename= 'West Bengal' GROUP BY district_name ORDER BY total_production DESC;

wb_rice = df[df["statename"] == "West Bengal"].groupby("district_name")["rice_production"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(y="rice_production", x="district_name", data=wb_rice, hue="district_name",legend=False,palette="Blues_r")
plt.ylabel("Rice Production")
plt.xlabel("District")
plt.title("Rice Production by West Bengal Districts")
plt.show()

```

```python
#Top 10 Wheat Production Years From UP
	#select year, SUM(wheat_production) AS total_production 
	#FROM AgriInfo WHERE statename= 'Uttar Pradesh' GROUP BY year ORDER BY total_production DESC limit 10;

up_wheat = df[df["statename"] == "Uttar Pradesh"].groupby("year")["wheat_production"].sum().reset_index()
up_wheat_top10 = up_wheat.nlargest(10, "wheat_production")

plt.figure(figsize=(10, 5))
sns.barplot(x="year", y="wheat_production", data=up_wheat_top10,hue="year",palette="Oranges_r")
plt.xlabel("Year")
plt.ylabel("Wheat Production")
plt.title("Top 10 Wheat Production Years from UP")
plt.show()
```

```python
#Millet Production (Last 50y)

	#select year,SUM(pearl_millet_production) AS total_production 
	#FROM AgriInfo  group by year ORDER BY total_production DESC limit 50;

Curryear = df["year"].max()
df_millet = df[df["year"] >= (Curryear - 50)]

millet_50y = df_millet.groupby("year")["pearl_millet_production"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x="year", y="pearl_millet_production", data=millet_50y, marker="o", color="purple", linewidth=2)
plt.xlabel("Year")
plt.ylabel("Millet Production")
plt.title("Millet Production Over Last 50 Years")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
#Sorghum Production (Kharif and Rabi) by Region
	#select statename,sum(kharif_sorghum_production) Kharif,sum(rabisorghum_production) Rabi from agriinfo
	#group by statename order by statename

sorghum_data = df.groupby("statename")[["kharif_sorghum_production", "rabisorghum_production"]].sum().reset_index()

# plot grouped bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="statename", y="kharif_sorghum_production", data=sorghum_data, color="orange", label="Kharif Sorghum")
sns.barplot(x="statename", y="rabisorghum_production", data=sorghum_data, color="blue", label="Rabi Sorghum")

plt.xlabel("State")
plt.ylabel("Production")
plt.title("Sorghum Production (Kharif & Rabi) by Region")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Top 7 States for Groundnut Production
	#select statename,sum(groundnut_production) Total_Production from agriinfo
	#group by statename order by Total_Production desc limit 7;
gtop7 = df.groupby("statename")["groundnut_production"].sum().nlargest(7).reset_index()

# Plot Horizontal Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x="groundnut_production", y="statename",hue="statename",legend=False, data=gtop7, palette="YlOrBr")
plt.xlabel("Groundnut Production")
plt.ylabel("State")
plt.title("Top 7 States for Groundnut Production")
plt.show()
```

```python
#Soybean Production by Top 5 States and Yield Efficiency
	#select statename,sum(soyabean_production) Total_Production,avg(soyabean_yield) Average_yield from agriinfo
	#group by statename order by Total_Production desc limit 5;

soybean_top5 = df.groupby("statename")[["soyabean_production", "soyabean_yield"]].sum().nlargest(5, "soyabean_production").reset_index()
fig, ax1 = plt.subplots(figsize=(10, 5))
sns.barplot(x="statename", y="soyabean_production", data=soybean_top5, ax=ax1, color="green", label="Soybean Production")
ax2 = ax1.twinx()
sns.lineplot(x="statename", y="soyabean_yield", data=soybean_top5, marker="o", color="blue", ax=ax2, label="Yield Efficiency")

# Labels and Title
ax1.set_xlabel("State")
ax1.set_ylabel("Soybean Production")
ax2.set_ylabel("Yield Efficiency (kg/ha)")
plt.title("Soybean Production and Yield Efficiency (Top 5 States)")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.show()

```

```python
#Oilseed Production in Major States

	#select statename,sum(oilseeds_production) Total_Production from agriinfo
	#group by statename order by Total_Production desc ;

oilseed_major = df.groupby("statename")["oilseeds_production"].sum().nlargest(10).reset_index()

# Treemap
plt.figure(figsize=(10, 6))
squarify.plot(sizes=oilseed_major["oilseeds_production"], label=oilseed_major["statename"], alpha=0.7, color=sns.color_palette("coolwarm"))
plt.title("Oilseed Production in Major States (Treemap)")
plt.axis("off")
plt.show()
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)
	#select year,sum(rice_area) ,sum(rice_production),
    #sum(wheat_area) ,sum(wheat_production),sum(maize_area),sum(maize_production) 
    #from Agriinfo group by year order by year ;

df_grouped = df.groupby("year")[["rice_area", "rice_production", "wheat_area", "wheat_production", "maize_area", "maize_production"]].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_grouped["rice_area"], y=df_grouped["rice_production"], label="Rice", color="blue", s=100)
sns.scatterplot(x=df_grouped["wheat_area"], y=df_grouped["wheat_production"], label="Wheat", color="red", s=100)
sns.scatterplot(x=df_grouped["maize_area"], y=df_grouped["maize_production"], label="Maize", color="green", s=100)
plt.xlabel("Area Cultivated (Hectares)")
plt.ylabel("Production")
plt.title("Impact of Area Cultivated on Production (Rice, Wheat, Maize)")
plt.legend()
plt.show()
    
```

```python
#Rice vs. Wheat Yield Across States

	#select statename,avg(rice_yield) ,avg(wheat_yield) 
    #from Agriinfo group by statename order by statename ;

yield_comparison = df.groupby("statename")[["rice_yield", "wheat_yield"]].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x="statename", y="rice_yield", data=yield_comparison, color="blue", label="Rice Yield")
sns.barplot(x="statename", y="wheat_yield", data=yield_comparison, color="red", alpha=0.7, label="Wheat Yield")

plt.xlabel("State")
plt.ylabel("Yield (kg/ha)")
plt.title("Rice vs. Wheat Yield Across States")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
